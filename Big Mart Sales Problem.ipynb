# Big Mart Sales Dataset

### Objective: To find out the properties of a product, and store which impacts the sales of a product.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
train = pd.read_csv("Big_Mart_Train.csv")
test = pd.read_csv("Big_Mart_test.csv")
```

```python
#Combine test and train into one file
data = pd.concat([train, test],ignore_index=True)
print(train.shape, test.shape, data.shape)
```

```python
#Numerical data summary:
data.describe()
```

```python
data.isnull().sum()
```

```python
data.info()
```

```python
#Replacing the missing value with mode (Categorical Data)
data['Outlet_Size'].fillna((data['Outlet_Size'].mode().values[0]), inplace=True)
```

```python
##Replacing the missing value with median
data['Item_Weight'].fillna(data['Item_Weight'].median(), inplace=True)
data['Item_Outlet_Sales'].fillna(data['Item_Outlet_Sales'].median(), inplace=True)
```

```python
data.isnull().sum()
```

```python
data.head(5)
```

```python
train.describe(include="all")
```

```python
#Item type combine:
data['Item_Identifier'].value_counts()
data['Item_Type_Combined'] = data['Item_Identifier'].apply(lambda x: x[0:2])
data['Item_Type_Combined'] = data['Item_Type_Combined'].map({'FD':'Food',
                                                             'NC':'Non-Consumable',
                                                             'DR':'Drinks'})
data['Item_Type_Combined'].value_counts()
```

```python
data.head(3)
```

```python
data.Item_Fat_Content.value_counts()
```

```python
data['Item_Fat_Content'] = data['Item_Fat_Content'].replace({'LF':'Low Fat',
                                                             'reg':'Regular',
                                                             'low fat':'Low Fat'})
print(data['Item_Fat_Content'].value_counts())
```

```python
data.head()
```

```python
#Drop the columns which have been converted to different types:
data.drop(['Item_Type','Outlet_Establishment_Year','Outlet_Identifier','Item_Identifier'],axis=1,inplace=True)
```

```python
data.head()
```

```python
data.drop(['Outlet_Identifier','Item_Identifier'],axis=1,inplace=True)
```

```python
data.head()
```

```python

data["Outlet_Type"] = data["Outlet_Type"].replace({'Grocery Store':0,'Supermarket Type1':1,'Supermarket Type2':2, 'Supermarket Type3':3})
print(data["Outlet_Type"].value_counts())
```

```python
data["Item_Fat_Content"] = data["Item_Fat_Content"].map({'Low Fat':0,'Regular':1,})
print(data["Item_Fat_Content"].value_counts())
```

```python
data.head()
```

```python
data["Outlet_Size"] = data["Outlet_Size"].replace({'Small':0,'Medium':1,'High':2})
print(data["Outlet_Size"].value_counts())
data["Outlet_Location_Type"] = data["Outlet_Location_Type"].replace({'Tier 1':0,'Tier 2':1,'Tier 3':2})
print(data["Outlet_Location_Type"].value_counts())
```

```python
data["Item_Type_Combined"] = data["Item_Type_Combined"].replace({'Food':0,'Drinks':1,'Non-Consumable':2})
print(data["Item_Type_Combined"].value_counts())
```

```python
data.head(6)
```

```python
X_new=data.loc[:,('Item_Weight','Item_Fat_Content','Item_Visibility','Item_MRP','Outlet_Size','Outlet_Location_Type','Outlet_Type','Item_Type_Combined')]
X_new.head()
Y=data.iloc[:,-2]
Y.head()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_val, y_train, y_val = train_test_split(X_new, Y, test_size = 0.25, random_state = 0)
```

```python
# Fitting Multiple Linear Regression to the training set
from sklearn.linear_model import  LinearRegression
regressor = LinearRegression()
regressor.fit(X_train, y_train)
```

```python
# Predicting the test set results
y_pred = regressor.predict(X_val)
y_pred
```

```python
# Import library for VIF
from statsmodels.stats.outliers_influence import variance_inflation_factor

def calc_vif(X_train):

    # Calculating VIF
    vif = pd.DataFrame()
    vif["variables"] = X_train.columns
    vif["VIF"] = [variance_inflation_factor(X_train.values, i) for i in range(X_train.shape[1])]

    return(vif)
calc_vif(X_train)
```
